```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_ds = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer_ds['data'],columns=cancer_ds['feature_names'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
#PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer_ds['target'])
plt.xlabel('First Principle Componenet')
plt.ylabel('Second Principle Componenet')
```

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer_ds['feature_names'])
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_comp)
```
